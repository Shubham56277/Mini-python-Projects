```python
from datetime import date

# current date
today = date.today()
cy, cm, cd = today.year, today.month, today.day

# birth date
by = int(input("enter year: "))
bm = int(input("enter month: "))
bd = int(input("enter day: "))

# days in months (index = month)
days_in_month = [31,28,31,30,31,30,31,31,30,31,30,31]

# age calculation
if cd < bd:
    cm -= 1
    cd += days_in_month[bm-1]

days = cd - bd

if cm < bm:
    cy -= 1
    cm += 12

months = cm - bm
years = cy - by

print("age is", years, "years", months, "months", days, "days")
```
